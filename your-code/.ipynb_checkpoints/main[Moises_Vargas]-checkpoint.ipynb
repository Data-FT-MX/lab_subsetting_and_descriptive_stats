# Challenge 1

#### In this challenge we will use the `Temp_States.csv` file. 

#### First import it into a data frame called `temp`.

```python
# your answer here
Temp=pd.read_csv('./Temp_States.csv')
```

#### Print `temp`

```python
Temp.head(10)
```

#### Explore the data types of the Temp dataframe. What type of data do we have? Comment your result.

```python
# your answer here
Temp.info()
Temp.columns
Temp['Temperature'].info()
```

#### Select the rows where state is New York

```python
# your answer here
Temp.iloc[0:3:]
```

#### What is the average of the temperature of cities in New York?

```python
# your answer here
Temp.iloc[0:3:].mean()
```

#### We want to know cities and states with Temperature above 15 degress Celcius

```python
# your answer here
Temp.loc[(Temp.loc[:, 'Temperature']>15)]
```

#### Now, return only the cities that have a temperature above 15 degress Celcius

```python
# your answer here
Temp.loc[(Temp.loc[:, 'Temperature']>15),['City']]
```

#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
# your answer here
# your answer here
Temp.loc[(Temp.loc[:, 'Temperature']>15) & (Temp['Temperature'] <20), ['City']]
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here
mean_temp= Temp.groupby('State').agg({'Temperature': 'mean'})
std_temp= Temp.groupby('State').agg({'Temperature': 'std'})
print(mean_temp)
print(std_temp)
```

#### Select orders from Spain that are above 50 items

```python
# your answer here
spain=orders.loc[(orders.loc[:, 'Country']=='Spain'),['CustomerID', 'Country', 'Quantity']]
spain_ID=spain.groupby('CustomerID').agg({'Quantity' : 'sum'})
spain_ID.loc[(spain_ID.loc[:,'Quantity']>50),'Quantity']
```
