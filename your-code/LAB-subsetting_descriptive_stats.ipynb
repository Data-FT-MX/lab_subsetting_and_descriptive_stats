#### Select the rows where state is New York

```python
# your answer here
temp[temp.State=='New York']
```

#### What is the average of the temperature of cities in New York?

```python
# your answer here
temp[temp.State == 'New York'].mean()
```

#### We want to know cities and states with Temperature above 15 degress Celcius

```python
temp[temp.Temperature>15]
#Df [Df.columna[condicion]]
```

#### Now, return only the cities that have a temperature above 15 degress Celcius

```python
# your answer here
temp.loc[:,['City', 'Temperature']][temp.Temperature>15]
#Df.loc[indice,[columnas]][Df.condicion] en la condicion especificar el Df al que se le aplicara
```

#### We want to know which cities have a temperature above 15 degrees Celcius and below 20 degrees Celcius

*Hint: First write the condition then select the rows.*

```python
# your answer here
temp[(temp.Temperature>15)&(temp.Temperature<20)].loc[:,['City', 'Temperature']]
# Df[(Df.columna <condicion 1>)&(Df.colummna <condicion 2>)].loc[:,[]]
```

#### Find the mean and the standard deviation of the temperature of each state.

*Hint: Use functions from Data Manipulation lesson*

```python
# your answer here

temp_mean = temp.groupby('State', as_index=False).mean()
temp_std_dev = temp.groupby('State', as_index=False).std()

temp_measures = pd.merge(temp_mean, temp_std_dev, on='State')
col= ['State', 'Mean', 'Stdev']
temp_measures.columns= col
temp_measures
```
